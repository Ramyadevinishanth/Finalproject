#Prediction on New Image

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt


# Load model
model = YOLO("/content/FruitsDetection-8/runs/detect/train/weights/best.pt")


# Load image
img_path = "/content/mixed_25.jpg"

img = cv2.imread(img_path)

if img is None:
    raise ValueError("Image not found. Check path!")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Run inference
results = model(img, conf=0.2,iou=0.5, imgsz=640)  # returns a list of Results

# Get first result

res = results[0]


# Visualize predictions
res.show()

# Print bounding boxes, class IDs, and confidences
if res.boxes is not None:
    boxes = res.boxes.xyxy.cpu().numpy()
    scores = res.boxes.conf.cpu().numpy()
    class_ids = res.boxes.cls.cpu().numpy()
    names = [res.names[int(i)] for i in class_ids]

    print("Detected objects:")
    for name, box, score in zip(names, boxes, scores):
        print(f"{name} | Confidence: {score:.2f} | Box: {box}")
else:
    print("No objects detected")

```
